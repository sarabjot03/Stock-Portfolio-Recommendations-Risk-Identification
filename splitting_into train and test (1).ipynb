Sort by Date and Symbol: Sort the data first by date in ascending order and then by symbol.(its already done but still in am making sure that it is done before the train and test split)
Split: Split the sorted data into a 70% training set and a 30% testing set.
Save as CSV: Save the resulting training and testing DataFrames to CSV files.

```python
df['date']=pd.to_datetime(df['date'])
df=df.sort_values(by=['date','symbol'])

```

```python
# Define the 70% split index
train_size = int(0.7 * len(df))
```

```python
# Split the data into training and testing sets
train_df = df[:train_size]
test_df = df[train_size:]
```

```python
 #Save the training and testing data to CSV files
train_df.to_csv(r"C:\Users\ASUS\Downloads\train_data.csv", index=False)
test_df.to_csv(r"C:\Users\ASUS\Downloads\test_data.csv", index=False)
```
